Есть предположение, что ```InCoder``` сможет легко справиться с классическими задачами, например, поиск факториала, бинарный поиск или сортировка массива. Однако попробовать стоит, но если дела будут совсем плохи, то можно попробовать решать классическую задачу с небольшими модификациями, например, в бинарный поиск добавить условие, что если элемента в массиве нет, то возвращать индекс по которому он мог бы там быть.

```>>> s = "cbbd"``` <br>```"bb```


Можно ли помочь модели найти правильное решение за счет подробного описания или тестов? Что-то одно из этого обязательно должно быть. Но если использовать только тесты, то это будет как черный ящик, это плохо потому что, во-первых, трудозатратно сидеть и придумывать тесты, которые бы покрыли всевозможные ситуации по задаче, а во-вторых, без описания на естественном языке даже людям было бы сложно вытащить идею про то о чем задача. Поэтому либо использовать только описание, либо описание вместе с тестами.

Загружем модель и импортируем метод, с помощью которого можно будет заполнять указанные места в коде

````python
from incoder import infill
````

Функция, которя принимает на вход докстринг и генерирует код

````python
def docstring_to_code(docstring: str, max_to_generate: int = 128, temperature: float = 0.2) -> str:
    example = f'''\
<insert>
    """{docstring}"""
    <insert>
'''
    
    result = infill(
        example.split("<insert>"), max_to_generate = max_to_generate, temperature = temperature
    )
    return result["text"]
````

Первое предположение. ```InCoder``` сможет легко сгенерировать код для классической задачи. Например, бинарный поиск.

````python
docstring = """Given a sorted array of distinct integers and a target value, return the index if the target is found"""
````

````python
%%time
from ipywidgets import IntProgress
from IPython.display import display

N = 10
snippets = []

progress = IntProgress(min=0, max=N)
display(progress)

for i in range(N):
    code = docstring_to_code(f'{docstring}')
    snippets.append(code)
    progress.value += 1
````

````python
for s in snippets:
    print(s, end=f"\n{'#'*100}\n")
````

````python
def binary_search(arr, target):
    
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        mid = (low + high)//2
        
        if arr[mid] == target:
            return mid
        
        elif arr[mid] > target:
            high = mid - 1
        
        else:
            low = mid + 1
    
    return -1
````

В этот раз реди сгенерированных сниппетов есть тот, который проходит минимальные тесты. 

````python
assert binary_search([1,3,5,7], 3) == 1
assert binary_search([1,3,5,7], 1) == 0
assert binary_search([1,3,5,7], 7) == 3

assert binary_search([1,3,5], 3) == 1
assert binary_search([1,3,5], 1) == 0
assert binary_search([1], 1) == 0
````

По возможности стоит указывать тесты, потому что без них модель не сразу понимает что от нее требуется. 

В целом можно сказать, что работающий код был получен довольно легко. Теперь можно усложнить, добавив условие, что если массив не содержит необходимый элемент, то нужно вернуть позицию куда этот элемент можно было бы вставить.

````python
assert binary_search([1,3,5,7], 4) == 2
assert binary_search([1,3,5,7], 0) == 0
assert binary_search([1,3,5,7], 8) == 4
````

Также ниже есть все функции, которые модель определила в качестве решений и их тестирование

````python
def longest_palindrome(s):
    """Given a string s, return the longest palindromic substring in s."""
    length = len(s)
    count = 0
    ans = ''
    for i in range(length):
        if s[i] == s[length - i - 1]:
            count += 1
        else:
            ans = s[i]
            break
    return ans, count
````

````python
assert longest_palindrome("babad") == "bab"
assert longest_palindrome("cbbd") == "bb"
````
